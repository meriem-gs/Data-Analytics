```python
import os
from xml.etree import ElementTree
```

```python
fileName = 'products.xml'
fullPath = os.path.abspath(os.path.join('',fileName))
print(fullPath)
```

```python
dom = ElementTree.parse(fullPath) 
print(dom)
```

```python
records = dom.findall('record/origin')
for c in records:
    print(c.text)
```
